 Sin el uso de librerías en Python programe el percentil y cuartil de cada columna.

```python
import pandas as pd

# Cargar el dataset
df = pd.read_csv("CALORIAS_DATASET.csv")

# Función para calcular percentiles sin librerías
def calcular_percentil(datos, percentil):
    datos_ordenados = sorted(datos)
    n = len(datos_ordenados)
    k = (n - 1) * (percentil / 100)
    f = int(k)
    c = k - f

    if f + 1 < n:
        resultado = datos_ordenados[f] + (c * (datos_ordenados[f + 1] - datos_ordenados[f]))
    else:
        resultado = datos_ordenados[f]

    return resultado

# Calcular cuartiles
def calcular_cuartiles(datos):
    q1 = calcular_percentil(datos, 25)
    q2 = calcular_percentil(datos, 50)  # Mediana
    q3 = calcular_percentil(datos, 75)
    return q1, q2, q3

# Calcular cuartiles para cada columna y presentarlos en apartados
def calcular_cuartiles_por_columna(df, columna):
    datos_columna = df[columna].dropna()  # Eliminar valores nulos

    # Convert the data to numeric type, handling errors
    try:
        datos_columna = pd.to_numeric(datos_columna)
    except ValueError:
        print(f"Error: La columna '{columna}' contiene valores no numéricos y no se puede calcular los cuartiles.")
        return

    q1, q2, q3 = calcular_cuartiles(datos_columna)
    print(f"========== Resultados para la columna: {columna} ==========")
    print(f"Percentil 25 (Q1): {q1}")
    print(f"Percentil 50 (Q2 - Mediana): {q2}")
    print(f"Percentil 75 (Q3): {q3}")
    print("=====================================================\n")

# Listado de columnas numéricas (ajústalo según tu dataset)
columnas_numericas = ['cal_por_100_ml_or_gms', 'kj_por_100_ml_or_gms', 'cal_por_servicio',
                      'kj_por_servicio', 'cal_por_onzas', 'kj_por_onzas']

# Calcular y mostrar los cuartiles para cada columna
for columna in columnas_numericas:
    calcular_cuartiles_por_columna(df, columna)
```

Que distribución se puede aplicar en su caso normal, Bernoulli, gaussiana, poisson, otros. Indique la razón de su uso graficando con matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Seleccionar una columna de ejemplo (calorías por 100 ml/gm)
datos = df['cal_por_100_ml_or_gms'].dropna()

# Graficar el histograma
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=30, color='lightblue', edgecolor='black', density=True, alpha=0.7)

# Ajustar una curva normal a los datos
media = np.mean(datos)
desviacion_estandar = np.std(datos)
x = np.linspace(min(datos), max(datos), 100)
y = (1 / (desviacion_estandar * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - media) / desviacion_estandar) ** 2)

# Graficar la curva normal
plt.plot(x, y, color='red', linewidth=2, label='Distribución Normal')

# Etiquetas y título
plt.title('Distribución de calorías por 100 ml/gm con ajuste normal')
plt.xlabel('Calorías por 100 ml/gm')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

El ajuste con la distribución normal no es perfecto debido a la asimetría de los datos. La curva normal no refleja adecuadamente la frecuencia de los valores más altos de calorías.

DISTRIBUCION LOG-NORMAL

```python
from scipy.stats import lognorm
import numpy as np
import matplotlib.pyplot as plt

# Seleccionar la columna de calorías por 100 ml/gm
datos = df['cal_por_100_ml_or_gms'].dropna()
#datos = df['kj_por_100_ml_or_gms'].dropna()
#datos = df['cal_por_servicio'].dropna()
#datos = df['kj_por_servicio'].dropna()
#datos = df['cal_por_onzas'].dropna()
#datos = df['kj_por_onzas'].dropna()

# Filtrar los valores 0
datos = datos[datos > 0]

# Graficar el histograma
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=30, color='lightblue', edgecolor='black', density=True, alpha=0.7)

# Ajustar una distribución log-normal a los datos
shape, loc, scale = lognorm.fit(datos, floc=0)  # Ajustar la distribución log-normal
x = np.linspace(min(datos), max(datos), 100)
pdf_lognorm = lognorm.pdf(x, shape, loc=loc, scale=scale)  # Función de densidad de probabilidad

# Graficar la curva log-normal ajustada
plt.plot(x, pdf_lognorm, color='green', linewidth=2, label='Distribución Log-normal')

# Etiquetas y título
plt.title('Distribución de calorías por 100 ml/gm con ajuste Log-normal')
plt.xlabel('Calorías por 100 ml/gm')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

La distribución log-normal proporciona un ajuste más preciso a los datos que la distribución normal. Captura mejor la naturaleza asimétrica de tus datos, que tienen muchos valores bajos y algunos valores extremos altos.
